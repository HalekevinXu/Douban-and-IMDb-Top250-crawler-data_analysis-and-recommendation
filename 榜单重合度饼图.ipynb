```python
from pyecharts.charts import Pie
from pyecharts import options as opts
import pandas as pd
import numpy as np

data1 = pd.read_csv('豆瓣电影.csv',encoding='gbk')
data2 = pd.read_csv('imdb_top_250_纯中文.csv',encoding='gbk')
data_db = data1['电影名称'].tolist()
data_im = data2['电影名称'].tolist()
same=0
db_in=0
for i in data_db:
    if i in data_im:
        same+=1
    else:
        db_in+=1
im_in=250-same
# print(same,db_in,im_in)
def pie_with_custom_label():
    pie = Pie(init_opts=opts.InitOpts(bg_color="white"))
    data = [("两榜同有", same), ("豆瓣特选", db_in), ("IMDb特选", im_in)]
    pie.add("", data)
    pie.set_global_opts(
        title_opts=opts.TitleOpts(
            title="榜单重合度饼图",
            title_textstyle_opts=opts.TextStyleOpts(color="#000000")
        )
    )
    # 自定义数据标签
    pie.set_series_opts(
        label_opts=opts.LabelOpts(
            position="outside",  # 标签位置在饼图外部
            formatter="{b}: {d}%", # 显示类别和百分比
            color="#333333"   # 深色文字
        )
    )
    return pie


chart = pie_with_custom_label()
chart.render_notebook()
# output_path = '榜单重合度饼图.html'
# chart.render(output_path)




    
```
